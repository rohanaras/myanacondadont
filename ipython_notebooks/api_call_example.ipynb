```python
import urllib
import json
import pandas as pd
```

```python
# read in json
url = 'https://www.nwfsc.noaa.gov/data/api/v1/source/trawl.catch_fact/selection.json?filters=' + \
      'date_dim$year%3E=2010,date_dim$year%3C=2012&variables=date_dim$yyyymmdd,cpue_kg_per_ha_der,' + \
      'cpue_numbers_per_ha_der'
api_response = urllib.urlopen(url)
json_file = json.loads(api_response.read())
```
